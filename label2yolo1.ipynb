```python
import os
import shutil
import random
from tqdm import tqdm
```

```python
os.chdir("../")
current_dir = os.getcwd()  # 获取当前工作目录
print("当前工作目录:", current_dir)
```

```python
Dataset_root = 'datasets/pose12'
```

```python
os.chdir(os.path.join(Dataset_root, 'images'))
os.listdir()
```

```python
# 划分训练集和测试集
test_frac = 0.2  # 测试集比例  20%
random.seed(123) # 随机数种子，便于复现
```

```python
folder = '.'  # 在当前目录下读取或写入文件，而不需要使用完整的路径
```

```python
img_paths = os.listdir(folder)
random.shuffle(img_paths) # 随机打乱

val_number = int(len(img_paths) * test_frac) # 测试集文件个数  25*0.2=5
train_files = img_paths[val_number:]         # 训练集文件名列表   前20个
val_files = img_paths[:val_number]           # 测试集文件名列表   后5个

print('数据集文件总数', len(img_paths))
print('训练集文件个数', len(train_files))
print('测试集文件个数', len(val_files))
```

```python
# 将训练集图像移动到images/trian目录
os.mkdir('train')   # 创建文件夹train
for each in tqdm(train_files):
    shutil.move(each, 'train')  # 转移
```

```python
# 将测试集图像移动到images/val目录
os.mkdir('val')
for each in tqdm(val_files):
    shutil.move(each, 'val')
```

```python
os.chdir('../labels')  # 改变当前工作目录
```

```python
# 将训练集标注移动到labelme_jsons/train目录
os.mkdir('train')
for each in tqdm(train_files):
    print(each)
    srt_path = each.split('.')[0] + '.json'   # 将 each 字符串中的文件名的扩展名修改为 .json
    shutil.move(srt_path, 'train')
```

```python
os.mkdir('val')
for each in tqdm(val_files):
    srt_path = each.split('.')[0] + '.json'
    shutil.move(srt_path, 'val')
```
